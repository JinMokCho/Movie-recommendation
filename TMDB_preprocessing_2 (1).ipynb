```python
import pandas as pd
from collections import Counter
data = pd.read_csv('C:/Users/carry/Documents/TMDB_158938')

# 'title_overview' 열을 만듭니다. 
data['title_overview'] = data['title'] + ' ' +data['overview']
data['title_overview'] = data['title_overview'].str.lower()

# 열을 삭제합니다.
data.drop(['Unnamed: 0','id','status','original_language','original_title','spoken_languages','overview'],axis = 1,inplace = True)
data.head()
```

```python
data.describe()
```

```python
# 영화 길이가 300분이 넘어가는 값을 삭제합니다.
data = data[data.runtime < 300].reset_index(drop = True)

#'Year'와'Month' 컬럼을 만듭니다.
data['year'] = data['release_date'].str[:4]
data['month'] = data['release_date'].str[5:7]
data.drop(['release_date'],axis = 1, inplace = True)

# 분석을 위해 'genres' 컬럼을 전처리합니다.
data['genres'] = data['genres'].str.split(',').apply(lambda x: [genre.strip().lower() for genre in x])
```

```python
# "genres"을 확인합니다.
genres_list = []
for i in range(len(data)):
    genres_list.extend(data.loc[i]['genres'])
print("All movie categories : ")
set(genres_list)
```

```python
# TOP_9 국가와 기타 국가로 분류합니다.
data['production_countries'] = data['production_countries'].str.split(',').apply(lambda x: [genre.strip().lower() for genre in x])
countries_list = []
for i in range(len(data)):
    countries_list.extend(data.loc[i]['production_countries'])
print('Top 9 countries with the highest occurrence :')
counter = Counter(countries_list)
top_9_countries = counter.most_common(9)
top_9_countries
```

```python
# List of TOP 9 countries
top_countries = [country for country, _ in top_9_countries]

# Change other countries name to 'Other'
data['production_countries'] = data['production_countries'].apply(lambda x: [country if country in top_countries else 'Other' for country in x])
```
